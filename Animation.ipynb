```python
import customtkinter as ctk
from PIL import Image
from os import walk

class AnimatedCheck(ctk.CTkFrame):
    def __init__(self, parent, image_path):
        super().__init__(master=parent, fg_color='white', width=100, height=100)
        self.grid_columnconfigure(0, weight=1)
        self.grid_rowconfigure(0, weight=1)
        self.grid(row=0, column=0)

        self.images = []  # This list will hold the CTkImage objects to prevent garbage collection
        self.load_images(image_path)
        self.display_first_image()

    def load_images(self, image_path):
        filenames = []
        for dirpath, dirnames, files in walk(image_path):
            for file in files:
                if file.endswith('.png') and not file.startswith('.DS_Store'):
                    filenames.append(file)

        sorted_filenames = sorted(filenames, key=lambda x: int(x.split('_')[1].split('.')[0]))

        for filename in sorted_filenames:
            full_path = f"{dirpath}/{filename}"
            image = Image.open(full_path)
            ctk_image = ctk.CTkImage(light_image=image, dark_image=image)
            self.images.append(ctk_image)

    def display_first_image(self):
        if self.images:
            label = ctk.CTkLabel(self, image=self.images[0])
            label.grid(row=0, column=0)

# window setup
window = ctk.CTk()
window.geometry('300x300')
window.title('Animation Display')

# Use a valid directory path for your images
image_path = 'path_to_your_image_directory'  # Replace with your actual image directory
AnimatedCheck(window, image_path)

window.mainloop()
```

### Resize images

```python
from PIL import Image
import os

def resize_image(image_path):
    # Load the image
    img = Image.open(image_path)
    
    # Get the directory and file name
    directory, filename = os.path.split(image_path)
    name, ext = os.path.splitext(filename)
    
    # Initialize the size and counter
    size = 3
    count = 1
    
    # Loop through the sizes from 3x3 to 60x60 in steps of 3
    while size <= 60:
        # Resize the image
        resized_img = img.resize((size, size), Image.NEAREST)
        
        # Save the resized image
        new_filename = f"{name}_{count}{ext}"
        new_filepath = os.path.join(directory, new_filename)
        resized_img.save(new_filepath)
        
        # Increment the size and count
        size += 3
        count += 1

# Example usage
resize_image('/Users/ps/Desktop/JIG 3000/checkmark/accept.png')
```
